```python
dir_list = os.listdir('.')
```

```python
txt_files = [x for x in dir_list if x.endswith("_timing.txt")]
```

```python
py_files = [x.replace('.txt', '.py') for x in txt_files]
```

```python
for i, j in enumerate(txt_files):
    with open(j,'r') as f:
        data = f.read()
    with open(py_files[i], 'w') as f:
        f.write(data)
```

```python
results = {}
```

```python
results['TS8004'] = ts.results
results['CY1003'] = cy1.results
results['CY11218'] = cy2.results
results['CY14025_c'] = cy3.results
results['CY14025_l'] = cy4.results
results['CY16204'] = cy5.results
results['CY9207'] = cy6.results
results['CY10206'] = cy7.results
```

```python
timing = {}
```

```python
timing['TS8004'] = ts.timing
timing['CY1003'] = cy1.timing
timing['CY11218'] = cy2.timing
timing['CY14025_c'] = cy3.timing
timing['CY14025_l'] = cy4.timing
timing['CY16204'] = cy5.timing
timing['CY9207'] = cy6.timing
timing['CY10206'] = cy7.timing
```

```python
def create_graph(results):
    Y = list(results.keys())
    X = list(results[Y[0]].keys())
    Z = []
    for x in Y:
        Z.extend(list(results[x].values()))
        
    x_ax, y_ax = np.meshgrid(X,Y)
    z_ar = np.asarray(Z)
    z_ax = z_ar.reshape(len(Y), len(X))
    
    plt.pcolormesh(x_ax,y_ax,z_ax, vmax=1)
    plt.colorbar()
    plt.show()
```

```python
create_graph(results)
```

```python
create_graph(timing)
```
